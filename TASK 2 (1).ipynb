```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.histplot(df['Selling_Price'], bins=10, kde=True, color='skyblue')
plt.title("Selling Price Distribution")
plt.xlabel("Selling Price (Lakhs)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Present_Price', y='Selling_Price', data=df, hue='Fuel_Type')
plt.title("Present Price vs Selling Price")
plt.xlabel("Present Price (Lakhs)")
plt.ylabel("Selling Price")
plt.show()

```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Fuel_Type', y='Selling_Price', data=df)
plt.title("Selling Price by Fuel Type")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Transmission', y='Selling_Price', data=df)
plt.title("Selling Price by Transmission Type")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.stripplot(x='Owner', y='Selling_Price', data=df, jitter=True, palette='Set2')
plt.title("Owner vs Selling Price")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Selling_type', data=df)
plt.title("Selling Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Fuel_Type', data=df)
plt.title("Fuel Type Distribution")
plt.show()
```
