```python
sns.histplot(df['Sales'], bins=10, kde=True)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.show()
```

```python
sns.scatterplot(x='TV', y='Sales', data=df)
plt.title("TV Budget vs Sales")
plt.show()
```

```python
sns.scatterplot(x='Radio', y='Sales', data=df)
plt.title("Radio Budget vs Sales")
plt.show()
```

```python
sns.scatterplot(x='Newspaper', y='Sales', data=df)
plt.title("Newspaper Budget vs Sales")
plt.show()
```

```python
sns.pairplot(df)
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(data=df)
plt.title("Boxplot of Features")
plt.show()
```

```python
sns.lineplot(data=df['TV'], marker='o')
plt.title("TV Budget Trend")
plt.xlabel("Index")
plt.ylabel("TV Budget")
plt.show()
```

```python
df['TV_HighLow'] = ['High' if x > df['TV'].median() else 'Low' for x in df['TV']]
sns.barplot(x='TV_HighLow', y='Sales', data=df)
plt.title("Average Sales by TV Spend Level")
plt.show()
```

```python
sns.kdeplot(df['Radio'], fill=True, label="Radio")
sns.kdeplot(df['Newspaper'], fill=True, label="Newspaper")
plt.title("KDE Plot of Radio and Newspaper Budget")
plt.legend()
plt.show()
```
